```python
evals = {'Micro':[
 'Epsilon',
 'Micro Precision',
 'Micro Recall',
 'Micro F1',
 ],

'Weighted and Macro Ancestors':[
 'Epsilon',
 'Weighted Precision for Ancestors',
 'Weighted Recall for Ancestors',
 'Weighted F1 for Ancestors',
 'Macro Precision for Ancestors',
 'Macro Recall for Ancestors',
 'Macro F1 for Ancestors',
 ],

'Weighted and Macro Descendants':[
 'Epsilon',
 'Weighted Precision for Descendants',
 'Weighted Recall for Descendants',
 'Weighted F1 for Descendants',
 'Macro Precision for Descendants',
 'Macro Recall for Descendants',
 'Macro F1 for Descendants',
 ],
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def visualize_scores(data, title, legend_loc, file_name, show_xticks=True):
  fig, ax = plt.subplots(figsize=(5, 4), dpi=200)
  x_column = 'Epsilon'
  x_values = data[x_column]
  palette = sns.color_palette("hls", len(data.columns)) #sns.color_palette("Paired", len(data.columns))

  for i, column in enumerate(data.columns):
      if column != x_column:
          ax.plot(x_values, data[column], marker='o', markersize=4,label=column, color=palette[i], linewidth=2)

  handles, labels = ax.get_legend_handles_labels()
  label_mapping = {
      'Micro Precision': 'Precision',
      'Micro Recall': 'Recall',
      'Micro F1': 'F1',
      'Weighted Precision for Ancestors': 'Weighted Precision',
      'Weighted Recall for Ancestors': 'Weighted Recall',
      'Weighted F1 for Ancestors': 'Weighted F1',
      'Macro Precision for Ancestors': 'Macro Precision',
      'Macro Recall for Ancestors': 'Macro Recall',
      'Macro F1 for Ancestors': 'Macro F1',
      'Weighted Precision for Descendants': 'Weighted Precision',
      'Weighted Recall for Descendants': 'Weighted Recall',
      'Weighted F1 for Descendants': 'Weighted F1',
      'Macro Precision for Descendants': 'Macro Precision',
      'Macro Recall for Descendants': 'Macro Recall',
      'Macro F1 for Descendants': 'Macro F1',
  }
  new_labels = [label_mapping.get(label, label) for label in labels]
  ax.legend(handles, new_labels, fontsize=12, loc=legend_loc)

  plt.xticks(x_values)
  plt.xticks(fontsize=14, rotation=90)
  plt.yticks(fontsize=16)

  if show_xticks == False:
    ax.set_xticks([])
    ax.set_xticklabels([])

  if show_xticks == True:
    ax.set_xlabel('Epsilon', fontsize=20)

  #ax.set_ylabel('Values', fontsize=16)
  #ax.set_title(title, fontsize=20)

  plt.show()
```

```python
show_xticks = True

for k, v in evals.items():
  filtered_df = thresholds_results_df
  filtered_df = filtered_df.rename(columns={'threshold': 'Epsilon'})
  filtered_metrics = filtered_df.filter(items=v)
  if k == 'Macro Descendants':
    show_xticks = True
  visualize_scores(filtered_metrics, r'$w^{Cui}$, ' + k, 'lower right', 'W Cui ' + k, show_xticks)
```
